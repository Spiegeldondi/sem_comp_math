```python
from ngsolve import *
from netgen.occ import unit_square
from ngsolve.webgui import Draw

# for adding interactive elements
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
# Generate unstructured mesh
mesh = Mesh(unit_square.GenerateMesh(maxh=0.1))
```

```python
# define a finite element space
fes_def = VectorH1(mesh, order=3)
# instantiate a grid function on the finite element space
gf_def = GridFunction(fes_def)
# deformation parameter
delta = Parameter(-0.3)
# define the grid function
gf_def.Set((0, delta*4*x*(1-x)*y))
# draw the grid function on the mesh
#Draw(gf_def)
```

```python
# apply grid function to mesh
mesh.SetDeformation(gf_def)
# draw the deformed mesh
scene = Draw(x, mesh)
```

```python
# wrap application of deformation in simple function
def deform(val=0.0):
    delta.Set(val)
    # update grid function
    gf_def.Set((0, delta*4*x*(1-x)*y))
    scene.Redraw()
```

```python
# create slider for updating deform() function
interact(deform, val=(-0.3, 0.4))
```
